```python
import pandas as pd
import numpy as np

import seaborn as sns
#import hvplot.pandas as plt
import matplotlib.pyplot as plt
import keras_tuner as kt

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

from sqlalchemy import create_engine, func

from Config import password

# Ignore warning messages
import warnings
warnings.filterwarnings("ignore")
```

```python
alchemy_engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/SolarDB')
```

```python
input_df = pd.read_sql_query('SELECT * FROM solar_data_v2;', alchemy_engine)
input_df
```

```python
input_df['Voltage_Output'] = input_df['InvVDCin_Avg']
input_df = input_df.drop(["TIMESTAMP", "InvVDCin_Avg"], axis=1)

input_df
```

```python
input_df.dtypes
```

```python
plt.scatter(input_df.Year, input_df.Voltage_Output)
plt.xlabel('Year')
plt.ylabel('Voltage_Output')
```

```python
y = input_df["Voltage_Output"].values
X = input_df.drop(["Voltage_Output"],1).values
input_features = (len(input_df.columns)-1)

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=27)
```

```python
# Initialize and fit the scaler object
scaler = StandardScaler()
```

```python
X_scaler = scaler.fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the basic neural network model
def create_model(hp):
    nn_model = tf.keras.models.Sequential()
    
    # Allow keras_tuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','elu','sigmoid'])
    
    # Allow keras_tuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(
        units=hp.Int('first_units',
                     min_value=6,
                     max_value=input_features,
                     step=10),
        activation='relu',
        input_dim=input_features))

    # Allow keras_tuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 3, 5)):
        nn_model.add(tf.keras.layers.Dense(
            units=hp.Int('units_' + str(i),
                         min_value=10,
                         max_value=40,
                         step=1),
            activation=activation))
    
    nn_model.add(tf.keras.layers.Dense(units=1, 
                                       activation="sigmoid"))

    # Compile the model
    nn_model.compile(loss = tf.keras.losses.MeanSquaredError(), 
                     optimizer="nadam", 
                     metrics=["accuracy"])
    
    return nn_model
    
```

```python
# Prepare the keras_tuner library
tuner = kt.Hyperband(
    create_model,
    objective="accuracy",
    max_epochs=20,
    hyperband_iterations=2,
    overwrite=True)
```

```python
# Run the keras_tuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,
             validation_data=(X_test_scaled,y_test))
```

```python
# Get top three model hyperparameters and print the values
top_hyper = tuner.get_best_hyperparameters(3)
for param in top_hyper:
    print(param.values)
```

```python
# Evaluate the top three models against the test dataset
top_model = tuner.get_best_models(3)
for model in top_model:
    model_loss, model_accuracy = model.evaluate(X_test_scaled,
                                                y_test,verbose=2)
    print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
