```python
import pandas as pd
```

```python
solar_df = pd.read_csv("../../Resources/Datasets/solar_data_predictions.csv")
solar_df = solar_df.drop(["Unnamed: 0"],1)
solar_df
```

```python
solar_df.columns
```

```python
solar_df['Array_Tilt'].value_counts()
```

```python
voltage_df = solar_df[['TIMESTAMP', 'Array_Tilt', 'Voltage_Predictions']]
voltage_df
```

```python
ground_mask = solar_df['Array_Tilt'] == 20
roof_mask = solar_df['Array_Tilt'] == 10
canopy_mask = solar_df['Array_Tilt'] == 5
```

```python
features_df = solar_df.drop(['Voltage_Output', 'Array_Tilt', 'Voltage_Predictions'],1)
features_df = features_df.drop_duplicates()
features_df = features_df.set_index('TIMESTAMP')

ground_df = solar_df[ground_mask]
ground_df = ground_df[['TIMESTAMP','Voltage_Output', 'Array_Tilt', 'Voltage_Predictions']]

roof_df = solar_df[roof_mask]
roof_df = roof_df[['TIMESTAMP', 'Voltage_Output', 'Array_Tilt', 'Voltage_Predictions']]

canopy_df = solar_df[canopy_mask]
canopy_df = canopy_df[['TIMESTAMP', 'Voltage_Output', 'Array_Tilt', 'Voltage_Predictions']]
```

```python
features_df
```

```python
df_1 = pd.merge(features_df, ground_df, on='TIMESTAMP', how='left')
df_1
```

```python
df_2 = pd.merge(df_1, roof_df, on='TIMESTAMP', how='left')
df_2
```

```python
df_3 = pd.merge(df_2, canopy_df, on='TIMESTAMP', how='left')
df_3 = df_3.set_index('TIMESTAMP')
df_3
```

```python
df_3.to_csv('sd_timestamps_predictions.csv')
```
