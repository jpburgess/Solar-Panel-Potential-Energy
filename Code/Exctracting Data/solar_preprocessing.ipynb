```python
t = "(2015)-(\d{2})-(\d{2})(\s)(\d{2}):(\d{2}):(\d{2})"
nist_2015 = nist_df[nist_df["TIMESTAMP"].str.contains(t)]
t = "(2016)-(\d{2})-(\d{2})(\s)(\d{2}):(\d{2}):(\d{2})"
nist_2016 = nist_df[nist_df["TIMESTAMP"].str.contains(t)]
t = "(2017)-(\d{2})-(\d{2})(\s)(\d{2}):(\d{2}):(\d{2})"
nist_2017 = nist_df[nist_df["TIMESTAMP"].str.contains(t)]
t = "(2018)-(\d{2})-(\d{2})(\s)(\d{2}):(\d{2}):(\d{2})"
nist_2018 = nist_df[nist_df["TIMESTAMP"].str.contains(t)]
```

```python
canopy_nist_2015 = nist_2015[nist_2015["Array Tilt"]==5]
roof_nist_2015 = nist_2015[nist_2015["Array Tilt"]==10]
ground_nist_2015 = nist_2015[nist_2015["Array Tilt"]==20]

canopy_nist_2016 = nist_2016[nist_2016["Array Tilt"]==5]
roof_nist_2016 = nist_2016[nist_2016["Array Tilt"]==10]
ground_nist_2016 = nist_2016[nist_2016["Array Tilt"]==20]

canopy_nist_2017 = nist_2017[nist_2017["Array Tilt"]==5]
roof_nist_2017 = nist_2017[nist_2017["Array Tilt"]==10]
ground_nist_2017 = nist_2017[nist_2017["Array Tilt"]==20]

canopy_nist_2018 = nist_2018[nist_2018["Array Tilt"]==5]
roof_nist_2018 = nist_2018[nist_2018["Array Tilt"]==10]
ground_nist_2018 = nist_2018[nist_2018["Array Tilt"]==20]
```

```python
solar_canopy_2015 = pd.merge(canopy_nist_2015, solar_nrel_2015, on='TIMESTAMP', how='left')
solar_canopy_2016 = pd.merge(canopy_nist_2016, solar_nrel_2016, on='TIMESTAMP', how='left')
solar_canopy_2017 = pd.merge(canopy_nist_2017, solar_nrel_2017, on='TIMESTAMP', how='left')
solar_canopy_2018 = pd.merge(canopy_nist_2018, solar_nrel_2018, on='TIMESTAMP', how='left')

solar_ground_2015 = pd.merge(ground_nist_2015, solar_nrel_2015, on='TIMESTAMP', how='left')
solar_ground_2016 = pd.merge(ground_nist_2016, solar_nrel_2016, on='TIMESTAMP', how='left')
solar_ground_2017 = pd.merge(ground_nist_2017, solar_nrel_2017, on='TIMESTAMP', how='left')
solar_ground_2018 = pd.merge(ground_nist_2018, solar_nrel_2018, on='TIMESTAMP', how='left')

solar_roof_2015 = pd.merge(roof_nist_2015, solar_nrel_2015, on='TIMESTAMP', how='left')
solar_roof_2016 = pd.merge(roof_nist_2016, solar_nrel_2016, on='TIMESTAMP', how='left')
solar_roof_2017 = pd.merge(roof_nist_2017, solar_nrel_2017, on='TIMESTAMP', how='left')
solar_roof_2018 = pd.merge(roof_nist_2018, solar_nrel_2018, on='TIMESTAMP', how='left')
```

```python
solar_canopy_2015 = solar_canopy_2015.dropna()
solar_canopy_2016 = solar_canopy_2016.dropna()
solar_canopy_2017 = solar_canopy_2017.dropna()
solar_canopy_2018 = solar_canopy_2018.dropna()

solar_ground_2015 = solar_ground_2015.dropna()
solar_ground_2016 = solar_ground_2016.dropna()
solar_ground_2017 = solar_ground_2017.dropna()
solar_ground_2018 = solar_ground_2018.dropna()

solar_roof_2015 = solar_roof_2015.dropna() 
solar_roof_2016 = solar_roof_2016.dropna() 
solar_roof_2017 = solar_roof_2017.dropna() 
solar_roof_2018 = solar_roof_2018.dropna()
```

```python
solar_canopy_2015.to_csv("solar_canopy_2015.csv", index = False)
solar_canopy_2016.to_csv("solar_canopy_2016.csv", index = False)
solar_canopy_2017.to_csv("solar_canopy_2017.csv", index = False)
solar_canopy_2018.to_csv("solar_canopy_2018.csv", index = False)

solar_ground_2015.to_csv("solar_ground_2015.csv", index = False)
solar_ground_2016.to_csv("solar_ground_2016.csv", index = False)
solar_ground_2017.to_csv("solar_ground_2017.csv", index = False)
solar_ground_2018.to_csv("solar_ground_2018.csv", index = False)

solar_roof_2015.to_csv("solar_roof_2015.csv", index = False)
solar_roof_2016.to_csv("solar_roof_2016.csv", index = False)
solar_roof_2017.to_csv("solar_roof_2017.csv", index = False)
solar_roof_2018.to_csv("solar_roof_2018.csv", index = False)
```
